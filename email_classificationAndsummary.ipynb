```python
import imaplib
import email
import chardet
from email.header import decode_header
import datetime
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from transformers import pipeline
```

```python

# 📌 Email Credentials
EMAIL_USER = "aryaidamle@gmail.com"
EMAIL_PASS = "bmpt iasw ikfb klrg"
IMAP_SERVER = "imap.gmail.com"

# 📌 Load Summarization Model
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

# 📌 Email Categories & Keywords
CATEGORY_KEYWORDS = {
    "Work/Business": ["meeting", "project", "deadline", "work", "business"],
    "Education & Learning": ["course", "lecture", "exam", "class"],
    "Finance & Banking": ["bank", "transaction", "payment", "loan"],
    "Legal & Government": ["law", "policy", "govt", "compliance"],
    "E-commerce & Orders": ["order", "invoice", "shipment", "delivery"],
    "Healthcare & Medical": ["doctor", "appointment", "prescription", "medicine"],
    "Technical Support": ["bug", "issue", "support", "ticket"],
    "Travel & Hospitality": ["flight", "hotel", "booking"],
    "Newsletters & Subscriptions": ["newsletter", "subscribe", "subscription"],
    "Social Media Notifications": ["Facebook", "Twitter", "Instagram"],
    "Personal": ["family", "friends", "birthday"],
    "Promotions": ["sale", "offer", "discount", "deal"]
}

# 📌 Connect to Email
def connect_email():
    """Connects to the email inbox securely."""
    try:
        mail = imaplib.IMAP4_SSL(IMAP_SERVER)
        mail.login(EMAIL_USER, EMAIL_PASS)
        mail.select("inbox")
        print("📩 Connected to email inbox!")
        return mail
    except Exception as e:
        print(f"❌ Error connecting to email: {str(e)}")
        return None

# 📌 Decode Text Safely
def decode_text(text_bytes):
    """Decodes unknown-encoding text safely."""
    if not text_bytes:
        return ""
    encoding = chardet.detect(text_bytes)["encoding"] or "utf-8"
    try:
        return text_bytes.decode(encoding, errors="ignore")
    except Exception:
        return text_bytes.decode("utf-8", errors="ignore")

# 📌 Fetch Emails Received Today
def fetch_today_emails(mail):
    """Fetches only today's emails."""
    try:
        today = datetime.datetime.today().strftime("%d-%b-%Y")  # Format as IMAP date
        status, messages = mail.search(None, f'(SENTON {today})')  # Fetch today's emails
        email_ids = messages[0].split()
        emails = []

        for e_id in email_ids:
            res, msg_data = mail.fetch(e_id, "(RFC822)")
            for response_part in msg_data:
                if isinstance(response_part, tuple):
                    msg = email.message_from_bytes(response_part[1])

                    # Decode subject
                    subject, encoding = decode_header(msg["Subject"])[0]
                    if isinstance(subject, bytes):
                        subject = decode_text(subject)

                    # Get sender & date
                    sender = msg.get("From", "Unknown Sender")
                    date = msg.get("Date", "Unknown Date")

                    # Extract email body
                    body = ""
                    if msg.is_multipart():
                        for part in msg.walk():
                            if part.get_content_type() == "text/plain":
                                body = decode_text(part.get_payload(decode=True))
                                break
                    else:
                        body = decode_text(msg.get_payload(decode=True))

                    emails.append({"date": date, "sender": sender, "subject": subject, "body": body})

        print(f"✅ Successfully fetched {len(emails)} emails from today!")
        return emails

    except Exception as e:
        print(f"⚠️ Error fetching emails: {str(e)}")
        return []

# 📌 Classify Email Based on Keywords
def classify_email(email_data):
    """Classifies an email based on its content."""
    for category, keywords in CATEGORY_KEYWORDS.items():
        if any(keyword.lower() in email_data["body"].lower() for keyword in keywords):
            return category
    return "Personal"

# 📌 Summarize Emails in Each Category
def summarize_category(email_texts):
    """Summarizes all emails in a category together."""
    combined_text = " ".join(email_texts)[:1024]  # Combine & truncate if too long
    inputs = tokenizer.encode("summarize: " + combined_text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=200, min_length=50, num_beams=4, early_stopping=True)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# 📌 Process, Classify & Summarize Emails
def process_emails():
    """Fetches, classifies, summarizes, and displays today's emails."""
    mail = connect_email()
    if not mail:
        return

    emails = fetch_today_emails(mail)
    if not emails:
        print("🚫 No emails received today.")
        return

    # 📌 Step 1: Classify Emails
    categorized_emails = defaultdict(list)
    for email_data in emails:
        category = classify_email(email_data)
        categorized_emails[category].append(email_data)

    # 📌 Step 2: Display Emails & Summarize Category-wise
    print("\n📊 **Classified Emails & Summaries**")
    for category, email_list in categorized_emails.items():
        print(f"\n🗂 **Category: {category}**")
        print("=" * 60)

        email_texts = []
        for email_data in email_list:
            print(f"📅 **Date:** {email_data['date']}")
            print(f"📧 **Sender:** {email_data['sender']}")
            print(f"📌 **Subject:** {email_data['subject']}")
            print("-" * 40)
            email_texts.append(email_data["body"])

        # 📌 Generate Summary for this Category
        if email_texts:
            summary = summarize_category(email_texts)
            print(f"\n📝 **Category Summary:** {summary}")
            print("=" * 60)
```

```python

# 📌 Run Every 10 Minutes
while True:
    process_emails()
    print("\n🔄 Checking again in 10 minutes...\n")
    time.sleep(600)  # Wait 10 minutes before rechecking
```
